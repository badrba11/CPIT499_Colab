#**ML Models**

##**Neural  Network Function**

```python
def NN(datap,iscross):
  x=datap.drop(columns=['State'])
  y=datap['State']
  X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=23)
  nn=MLPClassifier (activation="logistic", max_iter=500)
  nn.fit(X_train,y_train)
  Prediction=nn.predict(X_test)
  if(iscross):
      score = cross_val_score(nn , x , y , cv = 10) 
      return print('Ten folds Cross validation :{:.2f}'
      .format(sum(score)/len(score)))
  else: 
      score=accuracy_score(Prediction,y_test) 
      return print('Accuracy of NN classifier on test set: {:.2f}'
      .format(score)) 
```

```python
print('\nNormal data hot water with cross validation')

NN(hot,True)

print('\nNormal data cold water with cross validation')

NN(cold,True)

# without cross validation
print('\nNormal data hot water without cross validation')

NN(hot,False)
print('\nNormal data cold water without cross validation')

NN(cold,False)
```

##**Decision Tree Model Function**

```python
# Function to Tets Accurcy   
def DT(datap,maxdepth,iscross): 
   model=DecisionTreeClassifier(max_depth=maxdepth)
   x=datap.drop(columns=['State'])
   y=datap['State']
   X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.5, random_state=20)

   model.fit(X_train,y_train)
   Prediction=model.predict(X_test) 
   if(iscross):
      score = cross_val_score(model , x , y , cv = 10) 
      return print('Ten folds Cross validation :{:.2f}'
      .format(sum(score)/len(score)))
   else: 
      score=accuracy_score(Prediction,y_test) 
      return print('Accuracy of Decision Tree Model classifier on test set: {:.2f}'
      .format(score)) 
```

```python
print('Normal data hot water with cross validation')
DT(hot,5,True)

print('\nNormal data cold water with cross validation')
DT(cold,5,True)

print('\nNormal data hot water without cross validation')
DT(hot,5,False)

print('\nNormal data cold water without cross validation')
DT(cold,5,False)
```

##**K-Nearest Neighbors Function**

```python
def KNN(datap,iscross): 
   model=KNeighborsClassifier()
   x=datap.drop(columns=['State'])
   y=datap['State']
   X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.5, random_state=20)

   model.fit(X_train,y_train)
   Prediction=model.predict(X_test) 

   if(iscross):
      score = cross_val_score(model , x , y , cv = 10) 
      return print('Ten folds Cross validation :{:.2f}'
      .format(sum(score)/len(score))) 
   else:
      score=accuracy_score(Prediction,y_test) 
      return print('Accuracy of K-NN classifier on test set: {:.2f}'
     .format(score)) 
```

```python
print('Normal data hot water with cross validation')
KNN(hot,True)

print('Normal data cold water with cross validation')
KNN(cold,True)

print('Normal data hot water without cross validation')
KNN(hot,False)

print('Normal data cold water without cross validation')
KNN(cold,False)
```

##**Naive Bayes Function**

```python
def naive(datap,iscross): 
   gnb = GaussianNB() 
   x=datap.drop(columns=['State'])
   y=datap['State']
   X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=25)

   gnb.fit(X_train,y_train)

   Prediction=gnb.predict(X_test)   
  
   if(iscross):
      score = cross_val_score(gnb , x , y , cv = 10) 
      return print('Ten folds Cross validation  :{:.2f}'
      .format(sum(score)/len(score))) 
   else:
      score=accuracy_score(Prediction,y_test) 
      return print('Accuracy of Decision Tree classifier on test set is : {:.2f}'
      .format(score))
```

```python
print('Normal data hot water with cross validation')
naive(hot,True)

print('Normal data cold water with cross validation')
naive(cold,True)

print('Normal data hot water without cross validation')
naive(hot,False)

print('Normal data cold water without cross validation')
naive(cold,False)
```
